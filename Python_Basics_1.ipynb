1. What is Python, and why is it popular?

Python is a high-level, interpreted, and general-purpose programming language.
It is popular because of its simple syntax, readability, large standard library, cross-platform support, and wide applications in web development, data science, AI, automation, and embedded systems.

2. What is an interpreter in Python?

An interpreter is a program that executes Python code line by line.
Python does not require compilation before execution, which makes debugging easier and speeds up development.

3. What are pre-defined keywords in Python?

Pre-defined keywords are reserved words in Python that have special meanings and predefined functionality.
Examples: if, else, while, for, break, True, False, None, def, class.

4. Can keywords be used as variable names?

No, keywords cannot be used as variable names because they are reserved for specific operations in Python.
Using them as variables will result in a syntax error.

5. What is mutability in Python?

Mutability refers to the ability of an object to change its value after creation.
Mutable objects: list, dictionary, set
Immutable objects: tuple, string, integer

6. Why are lists mutable, but tuples are immutable?

Lists are mutable because they are designed for dynamic data modification.
Tuples are immutable to ensure data safety, faster execution, and fixed data storage.

7. What is the difference between “==” and “is” operators in Python?

== checks value equality
is checks object identity (memory location)

8. What are logical operators in Python?

Logical operators are used to combine conditional statements:

and returns True if both conditions are true
or returns True if at least one condition is true
not reverses the result of a condition

9. What is type casting in Python?

Type casting is the process of converting one data type into another.

10. What is the difference between implicit and explicit type casting?

Implicit type casting: Automatically done by Python (e.g., int to float)
Explicit type casting: Done manually by the programmer using functions like int(), float()

11. What is the purpose of conditional statements in Python?

Conditional statements allow a program to make decisions based on conditions and execute different blocks of code accordingly.
Examples: if, elif, else

12. How does the elif statement work?

elif is used to check multiple conditions after an if statement.
If the if condition is false, Python checks elif conditions in sequence until one is true.

13. What is the difference between for and while loops?

for loop: Used when the number of iterations is known

while loop: Used when the number of iterations is unknown and depends on a condition

14. Describe a scenario where a while loop is more suitable than a for loop.

A while loop is more suitable when user input or a condition controls repetition, such as:
Reading input until the user enters “exit”
Running a program until a sensor value reaches a threshold

```python
#13. Write a Python program to find the sum of all even numbers between 1 and 50
sum_even = 0

for i in range(1, 51):
    if i % 2 == 0:
        sum_even += i

print("Sum of even numbers between 1 and 50:", sum_even)
```

```python
#14. Write a program to reverse a string using a while loop
s = input("Enter a string: ")
rev = ""
i = len(s) - 1

while i >= 0:
    rev = rev + s[i]
    i = i - 1

print("Reversed string:", rev)
```

```python
#15. Write a Python program to calculate the factorial of a number provided by the user using a while loop
num = int(input("Enter a number: "))
fact = 1

while num > 0:
    fact = fact * num
    num = num - 1

print("Factorial:", fact)
```
