1. What is NumPy, and why is it widely used in Python?

NumPy is a Python library used for numerical and scientific computing.
It provides fast, efficient N-dimensional arrays and mathematical functions.
It is widely used because it is much faster than Python lists, supports vectorized operations, and is the foundation for libraries like Pandas, SciPy, and ML frameworks.

2. How does broadcasting work in NumPy?

Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically expanding the smaller array to match the larger one, without copying data, making operations efficient.

3. What is a Pandas DataFrame?

A DataFrame is a 2-dimensional labeled data structure in Pandas, similar to a table or Excel sheet, with rows and columns that can store different data types.

4. Explain the use of the groupby() method in Pandas

groupby() is used to split data into groups, apply aggregate functions (mean, sum, count), and then combine the results, commonly used in data analysis and summarization.

5. Why is Seaborn preferred for statistical visualizations?

6. What are the differences between NumPy arrays and Python lists
NumPy Array	Python List

7. What is a heatmap, and when should it be used?

8. What does “vectorized operation” mean in NumPy?

Vectorized operations perform calculations on entire arrays at once without explicit loops, making code faster and cleaner.

10. What is the Significance of hierarchical indexing in Pandas

Hierarchical indexing (MultiIndex) allows multiple levels of indexing, enabling complex data representation like higher-dimensional data in a 2D structure.

```python
#  How do you create a 2D NumPy array and calculate the sum of each row
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

row_sum = arr.sum(axis=1)
print(row_sum)
```

```python
#  Write a Pandas script to find the mean of a specific column in a DataFrame
import pandas as pd

data = {'Marks': [70, 80, 90, 85]}
df = pd.DataFrame(data)

mean_value = df['Marks'].mean()
print(mean_value)
```

```python
# Create a scatter plot using Matplotlib
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.scatter(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```

```python
# How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [4, 3, 2, 1],
    'C': [2, 3, 4, 5]
})

corr = df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
# Generate a bar plot using Plotly
import plotly.express as px

data = {'City': ['A', 'B', 'C'], 'Sales': [100, 200, 150]}
fig = px.bar(data, x='City', y='Sales', title="Bar Plot")
fig.show()
```

```python
# Create a DataFrame and add a new column based on an existing column
import pandas as pd

df = pd.DataFrame({'Salary': [30000, 40000, 50000]})
df['Bonus'] = df['Salary'] * 0.10

print(df)
```

```python
# Write a program to perform element-wise multiplication of two NumPy arrays
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = a * b
print(result)
```

```python
# Create a line plot with multiple lines using Matplotlib
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [15, 25, 35, 45]

plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2")
plt.legend()
plt.show()
```

```python
# Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold
import pandas as pd

df = pd.DataFrame({'Score': [50, 75, 90, 60]})
filtered = df[df['Score'] > 70]

print(filtered)
```

```python
# Create a histogram using Seaborn to visualize a distribution
import seaborn as sns
import matplotlib.pyplot as plt

data = [10, 20, 20, 30, 40, 40, 40, 50]
sns.histplot(data)
plt.show()
```

```python
# Perform matrix multiplication using NumPy
import numpy as np

a = np.array([[1, 2],
              [3, 4]])
b = np.array([[5, 6],
              [7, 8]])

result = np.dot(a, b)
print(result)
```
